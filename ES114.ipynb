```python

import matplotlib.pyplot as plt
from collections import Counter

def analyze_string(s):

    s = s.lower()
    char_counts = Counter(s)

    # Filter out non-alphabetic characters
    char_counts = {char: count for char, count in char_counts.items() if char.isalpha()}
    plt.figure(figsize=(10, 6))
    plt.bar(char_counts.keys(), char_counts.values(), color='skyblue')
    plt.title('Character Frequencies')
    plt.xlabel('Character')
    plt.ylabel('Frequency')
    plt.grid(True, axis='y', linestyle='--')
    plt.show()
text = "Hello World! I am Ankita Kushwaha."
frequencies = analyze_string(text)
print(frequencies)
```

```python
import statistics
numbers = [4, 6, 2, 4, 3, 6, 5, 6, 7, 3,9,4,5,2]

def remove_duplicates(lst):

    unique_elements = []
    for element in lst:
        if element not in unique_elements:
            unique_elements.append(element)
    if len(unique_elements) == 0:
        return None, None
    mean = sum(unique_elements) / len(unique_elements)
    median = statistics.median(unique_elements)
    return mean, median

mean, median = remove_duplicates(numbers)

print(f"Unique elements: {list(set(numbers))}")
print(f"Mean: {mean}")
print(f"Median: {median}")
```

```python

import matplotlib.pyplot as plt

def analyze_strings(strings):
    # Find the longest string
    longest_string = max(strings, key=len)

    # Count how many strings start with a vowel
    vowels = 'aeiouAEIOU'
    vowel_count = sum(1 for s in strings if s[0] in vowels)

    # Get the lengths of all strings
    string_lengths = [len(s) for s in strings]

    # Visualize the distribution of string lengths using a histogram
    plt.figure(figsize=(8, 6))
    plt.hist(string_lengths, bins=range(min(string_lengths), max(string_lengths) + 2), edgecolor='black', alpha=0.7)
    plt.title('Distribution of String Lengths')
    plt.xlabel('String Length')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return longest_string, vowel_count
strings = ["apple", "banana", "orange", "umbrella", "grape", "kiwi", "apricot"]
longest, vowels = analyze_strings(strings)

print(f"The longest string is: {longest}")
print(f"Number of strings that start with a vowel: {vowels}")
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
list1 = [1, 2, 3, 6, 5,8,9]
list2 = [4, 5, 6, 7, 8,5,7,2]
set1 = set(list1)
set2 = set(list2)
intersection = set1 & set2
union = set1 | set2
symmetric_difference = set1 ^ set2
print(f"Intersection: {intersection}")
print(f"Union: {union}")
print(f"Symmetric Difference: {symmetric_difference}")
plt.figure(figsize=(8, 6))
venn2(subsets=(len(set1 - set2), len(set2 - set1), len(intersection)),
      set_labels=('List 1', 'List 2'))
plt.title("Venn Diagram of Two Lists")
plt.show()
```
